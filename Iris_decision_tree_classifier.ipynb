#                          Task : Prediction using Decision Tree  Algorithm

We use the Decision Tree Classifier for the Iris dataset classification problem... And try to get maximum accuracy of our model...

```python
#-----import moduls---------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn import tree
from sklearn.metrics import accuracy_score
import warnings
warnings.simplefilter('ignore')
```

```python
#------------import the dataset---------------------
dataset = pd.read_csv('Iris.csv')
```

```python
#----------------------------checking for missing values-------------------------------------
dataset.isna().sum()
```

wow! its great, our dataset doesn't contain any missing value...

```python
#----------------------convert data in X and Y form---------------------------
X = dataset.iloc[:, 1:-1].values
Y = dataset.iloc[:, -1].values
```

```python
#-----------------------Encoding the dependent variable------------------------
from sklearn.preprocessing import LabelEncoder
lable = LabelEncoder()
Y = lable.fit_transform(Y)
```

```python
#------------make partition of train, test dataset-------------------------
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.20,random_state=0)
```

```python
#------------------Let's apply Desicion tree classification model-----------------
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='gini')
dt.fit(X_train,Y_train)
Y_pred = dt.predict(X_test)
print(Y_pred)
```

```python
#-------------Let check the accuracy of our model------------------------
acc_dt = accuracy_score(Y_test, Y_pred)
cm_dt = confusion_matrix(Y_test,Y_pred)
print("Accuracy : "+str(acc_dt))
print("Confusion_matrix :"+str(cm_dt))
```

Wow! Our accuracy is 100%, And that's becuase our dataset is small and probability will increase to get 100% accuracy by using DecisionTreeClassifier Model...

```python
#-------------------------------Let's Visualize the Train-Data----------------------------
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, Y_train
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 2],
                c = ListedColormap(('red','blue','green'))(i), label = j)
plt.title('Decision Tree Classification (Training set)')
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.legend()
plt.show()
```

```python
#-------------------------------Let's Visualize the Test-Data----------------------------
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, Y_test
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 2],
                c = ListedColormap(('red','blue','green'))(i), label = j)
plt.title('Decision Tree Classification (Test set)')
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.legend()
plt.show()
```

```python
#------------Let's visualize the trees--------------------------------
fig = plt.figure(figsize=(25,20))
fig = tree.plot_tree(dt,filled=True, feature_names=dataset.iloc[:, 1:-1].columns)
```
